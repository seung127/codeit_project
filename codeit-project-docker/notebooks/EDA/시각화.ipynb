```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors

#   
total_dau["date"] = pd.to_datetime(total_dau["date"])

# Normalize: DAU    0~1 
norm = mcolors.Normalize(vmin=total_dau["DAU"].min(), vmax=total_dau["DAU"].max())

# Colormap: Oranges   
cmap = cm.get_cmap("Oranges")

# DAU     
colors = [cmap(norm(dau)) for dau in total_dau["DAU"]]

#  
plt.figure(figsize=(12, 6))
plt.bar(
    total_dau["date"],
    total_dau["DAU"],
    color=colors
)
plt.title("DAU  (7 13 - 7 17  )", fontsize=18)
plt.xlabel("", fontsize=17)
plt.ylabel("   ", fontsize=17)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#  
df = total_wau.copy()

# Normalize: WAU   
norm = mcolors.Normalize(vmin=df["WAU"].min(), vmax=df["WAU"].max())

# Colormap: Oranges 
cmap = cm.get_cmap("Oranges")

#   
colors = [cmap(norm(wau)) for wau in df["WAU"]]

# 
plt.figure(figsize=(12, 6))
plt.bar(df["week"], df["WAU"], color=colors)
plt.title("WAU", fontsize=18)
plt.xlabel("", fontsize=17)
plt.ylabel("  ", fontsize=17)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 
df = rolling_wau.copy()

# Normalize:    ( )
norm = mcolors.Normalize(vmin=df["rolling WAU"].min(), vmax=df["rolling WAU"].max())
cmap = cm.get_cmap("Oranges")
colors = [cmap(norm(val)) for val in df["rolling WAU"]]

# 
plt.figure(figsize=(12, 6))
plt.bar(df["date"], df["rolling WAU"], color=colors)

# 
plt.title("7 Rolling WAU (Weekly Active Users)", fontsize=18)
plt.xlabel("", fontsize=14)
plt.ylabel("  ", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
