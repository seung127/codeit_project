```python
# 날짜 형식 변환
total_dau["date"] = pd.to_datetime(total_dau["date"])

# 기준 설정: 상위 20% 이상은 주황
threshold = total_dau["DAU"].quantile(0.80)
colors = ["#fca311" if val >= threshold else "#264653" for val in total_dau["DAU"]]

# 그래프
plt.figure(figsize=(12, 6))
plt.bar(
    total_dau["date"],
    total_dau["DAU"],
    color=colors
)
plt.title("DAU 추이", fontsize=18)
plt.xlabel("날짜", fontsize=17)
plt.ylabel("일간 활성 사용자 수", fontsize=17)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors

# 날짜 형식 변환
total_dau["date"] = pd.to_datetime(total_dau["date"])

# Normalize: DAU 값의 상대 크기를 0~1로 정규화
norm = mcolors.Normalize(vmin=total_dau["DAU"].min(), vmax=total_dau["DAU"].max())

# Colormap: Oranges 컬러맵에서 색상 추출
cmap = cm.get_cmap("Oranges")

# DAU 값에 따라 색상 리스트 생성
colors = [cmap(norm(dau)) for dau in total_dau["DAU"]]

# 막대그래프 그리기
plt.figure(figsize=(12, 6))
plt.bar(
    total_dau["date"],
    total_dau["DAU"],
    color=colors
)
plt.title("DAU 추이 (7월 13일 - 7월 17일 데이터 없음)", fontsize=18)
plt.xlabel("날짜", fontsize=17)
plt.ylabel("일간 활성 사용자 수", fontsize=17)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 복사본 생성
df = total_wau.copy()

# Normalize: WAU 값 기준 정규화
norm = mcolors.Normalize(vmin=df["WAU"].min(), vmax=df["WAU"].max())

# Colormap: Oranges 팔레트
cmap = cm.get_cmap("Oranges")

# 색상 리스트 생성
colors = [cmap(norm(wau)) for wau in df["WAU"]]

# 그래프
plt.figure(figsize=(12, 6))
plt.bar(df["week"], df["WAU"], color=colors)
plt.title("WAU", fontsize=18)
plt.xlabel("주차", fontsize=17)
plt.ylabel("고유 사용자 수", fontsize=17)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 복사
df = rolling_wau.copy()

# Normalize: 값 기반 정규화 (높을수록 진한색)
norm = mcolors.Normalize(vmin=df["rolling WAU"].min(), vmax=df["rolling WAU"].max())
cmap = cm.get_cmap("Oranges")
colors = [cmap(norm(val)) for val in df["rolling WAU"]]

# 그래프
plt.figure(figsize=(12, 6))
plt.bar(df["date"], df["rolling WAU"], color=colors)

# 서식
plt.title("7일 Rolling WAU (Weekly Active Users)", fontsize=18)
plt.xlabel("날짜", fontsize=14)
plt.ylabel("고유 사용자 수", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
mau = total_mau.copy()

# 그래프 설정
plt.figure(figsize=(12, 6))
bars = plt.bar(
    mau["month"].astype(str),       # 월을 문자열로 변환하여 x축 라벨로
    mau["MAU"],
    color="#fca311",                # 연초록          # 테두리 초록
    width=0.6
)

# 타이틀 및 축 서식
plt.title("MAU", fontsize=18)
plt.xlabel("월", fontsize=17)
plt.ylabel("고유 사용자 수", fontsize=17)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# 데이터 복사
mau = total_mau.copy()

# Normalize: MAU 값 기준으로 정규화 (높을수록 진한 색)
norm = mcolors.Normalize(vmin=mau["MAU"].min(), vmax=mau["MAU"].max())
cmap = cm.get_cmap("Oranges")
colors = [cmap(norm(val)) for val in mau["MAU"]]

# 그래프 그리기
plt.figure(figsize=(12, 6))
bars = plt.bar(
    mau["month"].astype(str),  # 문자열로 변환
    mau["MAU"],
    color=colors,
    width=0.6
)

# 서식
plt.title("MAU", fontsize=18)
plt.xlabel("월", fontsize=17)
plt.ylabel("고유 사용자 수", fontsize=17)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```

```python
# 데이터 복사
mau = total_mau.copy()

# 최고값 인덱스
max_index = mau["MAU"].idxmax()

# 색상 리스트 생성: 최고값은 주황, 나머지는 남색
colors = [
    "#fca311" if i == max_index else "#14213d"  # 남색: #264653
    for i in range(len(mau))
]

# 그래프
plt.figure(figsize=(12, 6))
bars = plt.bar(
    mau["month"].astype(str),
    mau["MAU"],
    color=colors,
    width=0.6
)

# 서식
plt.title("MAU", fontsize=18)
plt.xlabel("월", fontsize=17)
plt.ylabel("고유 사용자 수", fontsize=17)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.tight_layout()
plt.show()
```
