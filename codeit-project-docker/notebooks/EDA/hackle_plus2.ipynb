## 💡 아이디어: `session_start` 데이터만으로 분석하기

`$session_start` 이벤트만을 활용해도 아래와 같은 인사이트 도출이 가능함:

1. **유저의 활동 시간대 분석**
   - `session_start` 발생 시각을 기반으로
   - 시간대별 접속 분포 파악 (예: 오전/오후/야간 이용량)
   - 특정 시간대에 사용자가 몰리는지 확인 가능

2. **앱의 이용량 측정**
   - `session_start`의 발생 횟수 = 앱을 켠 횟수
   - 일/주/월 단위로 집계하여 **DAU/WAU/MAU** 대체 지표로 활용 가능
   - 기능/이벤트별 영향 분석 시 baseline 지표로 사용 가능

```python
date_list = [df_230430, df_230506, df_230512, df_230518, df_230524, df_230530]
date_names = ['230430', '230506', '230512' ,'230518', '230524', '230530']

for name, data in zip(date_names, date_list):
    print(f"📅 날짜: {name}")
    
    df = data.groupby('session_id')['device_id'].size().reset_index(name='event_count')
    display(df['event_count'].value_counts().sort_index())
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

# 1. Style setting
sns.set(style='whitegrid')
plt.rcParams['axes.unicode_minus'] = False  # Ensure minus signs display correctly

# 2. Plot session event count distribution per date
for name, data in zip(date_names, date_list):
    # Count number of events per session
    session_event_counts = data.groupby('session_id').size().reset_index(name='event_count')

    # Get frequency distribution of event counts
    event_distribution = session_event_counts['event_count'].value_counts().sort_index()
    event_distribution = event_distribution[event_distribution.index >= 1]

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.bar(event_distribution.index, event_distribution.values, width=0.6)

    plt.title(f'Session Event Count Distribution - {name}', fontsize=14)
    plt.xlabel('Number of Events per Session', fontsize=12)
    plt.ylabel('Number of Sessions', fontsize=12)
    plt.xticks(event_distribution.index, rotation=0)
    plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
    plt.tight_layout()
    plt.show()
```
