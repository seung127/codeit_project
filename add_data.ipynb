```python
import os
import pandas as pd

# CSV 파일이 저장된 디렉토리 경로
data_dir = '/home/hq_key/codeit-project-docker/data/'

# 해당 폴더 안의 모든 파일 이름을 가져오고, csv만 필터링
parquet_files = sorted([
    file for file in os.listdir(data_dir) 
    if file.endswith('.parquet') and file.startswith('23')
])

# 각 파일을 읽어서 리스트에 저장
dataframes = []
for file in parquet_files:
    file_path = os.path.join(data_dir, file)
    df = pd.read_parquet(file_path)
    dataframes.append(df)
    print(f"{file} 로드 완료, shape: {df.shape}")
```

```python
df_430 = dataframes[0]
df_506 = dataframes[1]
df_512 = dataframes[2]          
df_518 = dataframes[3]
df_524 = dataframes[4]
df_530 = dataframes[5]
df_605 = dataframes[6]
df_611 = dataframes[7]
df_617 = dataframes[8]
```
