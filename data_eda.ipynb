```python
language_counts = hackle_properties['language'].value_counts()
language_percent = (language_counts / language_counts.sum()) * 100

# 상위 5 언어만
top_5 = 5
top_languages = language_percent[:top_5]
other_percent = 100 - top_languages.sum()

# 파이차트용 라벨 + 기타 묶기
labels = list(top_languages.index) + ['기타']
sizes = list(top_languages.values) + [other_percent]

# 시각화
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('언어별 사용자 비율')

plt.show()

# 비율 출력
print(" 언어별 사용자 비율 상위 10:")
print(top_languages.round(2).astype(str) + ' %')
```
