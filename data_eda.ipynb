```python
# 예시: 두 테이블이 school과 user라고 가정
school_ids_1 = set(school['id'])
school_ids_2 = set(user_properties['school_id'])

# 교집합 (중첩되는 school_id)
common_ids = school_ids_1 & school_ids_2

# 차집합 (school 테이블에만 있는 school_id)
only_in_school = school_ids_1 - school_ids_2

# 차집합 (user 테이블에만 있는 school_id)
only_in_user = school_ids_2 - school_ids_1

# 출력
print(f"중첩되는 school_id 개수: {len(common_ids)}")
print(f"school 테이블에만 있는 school_id 개수: {len(only_in_school)}")
print(f"user 테이블에만 있는 school_id 개수: {len(only_in_user)}")
# 그렇다면 2개의 school_id 예외 사항을 제외한 나머지 user_id에는 user_properties를 기준으로 user_id에 어떤 지역 학생인지 엮을 수 있다.
```

```python
hackle_events['friend_bin'] = pd.cut(hackle_events['friend_count'], bins=[0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700],)
hackle_events['friend_bin'].value_counts()

avg_votes_per_bin = hackle_events.groupby('friend_bin')['votes_count'].mean().sort_index()


avg_votes_per_bin.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('친구 수 구간별 평균 투표 수', fontproperties=fontprop)
plt.xlabel('Friend Count 구간', fontproperties=fontprop)
plt.ylabel('Average Votes Count', fontproperties=fontprop)
plt.xticks(rotation=45, fontproperties=fontprop)
plt.tight_layout()
plt.show()
```
