```python
import pandas as pd

# 사전 정렬
df = final_hackle[['user_id', 'event_key', 'event_datetime', 'heart_balance']].copy()
df = df.sort_values(['user_id', 'event_datetime'])

# 이전 하트 잔액, 이벤트 키, 시간 계산
df['heart_balance_prev'] = df.groupby('user_id')['heart_balance'].shift(1)
df['event_key_prev'] = df.groupby('user_id')['event_key'].shift(1)
df['event_datetime_prev'] = df.groupby('user_id')['event_datetime'].shift(1)

# 변화량 계산 공식
df['heart_balance_diff'] = df['heart_balance'] - df['heart_balance_prev']

# NaN 이전 값을 채우기
df['heart_balance'] = df['heart_balance'].fillna(method='ffill')

# NaN 제거 & 변화가 있는 경우만 골라내기
result_df = df[
    df['heart_balance_diff'].notnull() &
    (df['heart_balance_diff'] != 0)
].copy()



# 원하는 컬럼만 정리
result_df = result_df[[
    'user_id',
    'event_key_prev', 'event_datetime_prev', 'heart_balance_prev',
    'event_key', 'event_datetime', 'heart_balance',
    'heart_balance_diff'
]]

result_df
```

### [3] OS 버전 변화와 `session_id` 연관성 검토 (민정,연우)

- **질문**: 사용자의 OS 버전(`os_version`)이 바뀌면 `session_id`도 바뀌는가?
- **분석 방향**:
    - 동일 유저 내 시간 순으로 `os_version`과 `session_id` 변화 패턴 분석.
    - 특정 버전 변경이 `session_id` 변경과 직접적인 관련이 있는지 여부 확인.

```python
import pandas as pd

# 데이터 불러오기
df = hackle_properties.copy()

# osversion 변경 감지를 위해 정렬
df = df.sort_values(by=['user_id', 'id'])  # 시간순 정렬 (id 기준)

# 사용자별 osversion 이전 값과 비교
df['osversion_prev'] = df.groupby('user_id')['osversion'].shift(1)
df['session_id_prev'] = df.groupby('user_id')['session_id'].shift(1)

# osversion이 변경된 행만 필터링
os_changed = df[df['osversion'] != df['osversion_prev']].copy()

# 각 행에 대해 session_id가 변경되었는지 여부 계산
os_changed['session_changed'] = os_changed['session_id'] != os_changed['session_id_prev']

# 결과 확인
summary = os_changed['session_changed'].value_counts()

# 어떤 user_id에서 변경이 있었는지 보기
changed_users = os_changed[['user_id', 'osversion_prev', 'osversion', 'session_id_prev', 'session_id', 'session_changed']]
# 해클 프로퍼티스랑 이벤트 엮고 session_id 날리고 분석
```

```python
import pandas as pd

df = hackle_properties.copy()

# 시간 순 정렬
df = df.sort_values(by=['user_id', 'id'])  # 또는 timestamp가 있다면 그걸 기준으로

# 이전 osversion과 비교
df['osversion_prev'] = df.groupby('user_id')['osversion'].shift(1)

# osversion이 바뀐 행만 필터링
df['osversion_changed'] = df['osversion'] != df['osversion_prev']

# osversion 변경이 있었던 user_id 집합 추출
user_with_os_change = df[df['osversion_changed']]['user_id'].unique()

# 전체 사용자 수
total_users = df['user_id'].nunique()

# 변경된 사용자 수
changed_users = len(user_with_os_change)

# 비율 계산
change_ratio = changed_users / total_users * 100

print(f"전체 사용자 수: {total_users}")
print(f"osversion이 변경된 사용자 수: {changed_users}")
print(f"변경된 사용자 비율: {change_ratio:.2f}%")
```
